```python
import pandas as pd 
data = pd.read_csv('gbm-data.csv')
data
```

```python
import numpy as np
x = np.array(data.values[:,1:])
y = np.array(data.values[:,0])

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.8, random_state=241)
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import log_loss
lrs = [1, 0.5, 0.3, 0.2, 0.1]
results = {}
for lr in lrs:
    clf = GradientBoostingClassifier(n_estimators=250, learning_rate=lr,
                                 verbose=True, random_state=241)
    clf.fit(x_train, y_train)
    
    train_prediction = np.array([x for x in clf.staged_decision_function(x_train)]).reshape(250,750)
    train_prediction = sigmoid(train_prediction)
    
    test_prediction = np.array([x for x in clf.staged_decision_function(x_test)]).reshape(250,3001)
    test_prediction = sigmoid(test_prediction)
    
    train_score = []
    test_score = []
    for it in train_prediction:
        score = log_loss(y_train, it)
        train_score.append(score)
        
    for it in test_prediction:    
        score = log_loss(y_test, it)
        test_score.append(score)
    
    results[lr] = {"train":train_score, "test":test_score}
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure()
plt.plot(results[1]['train'],'r-', linewidth=2)
plt.plot(results[1]['test'],'g-', linewidth=2)

plt.plot(results[0.5]['train'],'ro', linewidth=2)
plt.plot(results[0.5]['test'],'go', linewidth=2)

plt.plot(results[0.3]['train'],'rx', linewidth=2)
plt.plot(results[0.3]['test'],'gx', linewidth=2)

plt.plot(results[0.2]['train'],'r+', linewidth=2)
plt.plot(results[0.2]['test'],'g+', linewidth=2)

plt.plot(results[0.1]['train'],'r.', linewidth=2)
plt.plot(results[0.1]['test'],'g.', linewidth=2)

plt.legend(['train-1','test-1','train-0.5','test-0.5','train-0.3','test-0.3',
           'train-0.2','test-0.2','train-0.1','test-0.1'])
```
